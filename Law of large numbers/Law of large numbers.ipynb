# Law of large numbers

## Formula

* The law suggests that as the number of trials increase, the sample average gets closer to it's expected value.

* Let $X$ be a discrete random variable. The formula is given as

$$
\begin{align}
\mathbb{E}[X] = \sum_{i=1}^{n} x_i P(x_i) \\
\end{align}
$$

where the first term represents a value and the latter its probability.

## Example calculation

* For example, let's call $X$ the obtained number when rolling a six sided dice. Therefore, its expected value is

$$
\begin{align}
\mathbb{E}[X] = \frac{1}{6}(1+2+3+4+5+6)=3.5 \\
\end{align}
$$

# Simulation

```python
import random
import matplotlib.pyplot as plt
```

```python
# Create lists to store the samples and their averages
samples = []
averages = []
```

```python
# Number of simulations
num_simulations = 5000

# Sides on the dice
num_sides = 6

# Simulate the rolls
for i in range(num_simulations):
    roll = random.randint(1, num_sides)
    samples.append(roll)

for i in range(num_simulations):
    sample = samples[0:i+1]
    averages.append(sum(sample)/len(sample))
```

```python
plt.plot(averages)
plt.ylabel('Sample average')
plt.xlabel('Trial')
plt.axhline(y = 3.5, color = 'g', label = "Expected value")
plt.legend()

plt.show()
```

* It can clearly be seen that the sample average tends to get closer to the expected value of $X$.
